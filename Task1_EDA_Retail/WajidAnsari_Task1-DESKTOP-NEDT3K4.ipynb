```python
import pandas as pd
df = pd.read_csv('retail_sales.csv')
df.head()

```

```python
print("Total rows, columns:", df.shape)
print("Duplicated rows:", df.duplicated().sum())
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
df.describe()
```

```python
print("Branch wise billing count:")
print(df['Branch'].value_counts())
```

```python
print("product line wise count:")
print(df['Product line'].value_counts())
```

```python
print("Before conversion:", df['Date'].dtype)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce') 
print("After conversion:", df['Date'].dtype)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df.groupby('Branch')['Sales'].sum().plot(kind='bar', color=['#FF9999', '#66B2FF', '#99FF99'])
plt.title('Branch Wise Total Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df.groupby('Product line')['Sales'].sum().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Product Line Wise Total Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Payment'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightblue'])
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.show()
```

1.**Branch C** has highest sales compared to Branch A and B
2.**Food & beverages** is top selling product line
3.**Ewallet & Cash** are most used payment methods ~34% each
**Recommendation**: Focus more on Branch C strategies for other branches
