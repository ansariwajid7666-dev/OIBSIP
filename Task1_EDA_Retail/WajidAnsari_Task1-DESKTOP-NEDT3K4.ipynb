```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df.groupby('Branch')['Sales'].sum().plot(kind='bar', color=['#FF9999', '#66B2FF', '#99FF99'])
plt.title('Branch Wise Total Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df.groupby('Product line')['Sales'].sum().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Product Line Wise Total Sales')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Payment'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightblue'])
plt.title('Payment Method Distribution')
plt.ylabel('')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Day 5: Correlation Heatmap')
plt.show()
```

1. **Perfect Correlation (1.0):** Sales, Tax 5%, and cogs have a perfect positive correlation.
2. **Strong Correlation (0.71):** Quantity shows a strong positive correlation with sales.
3. **No Correlation (~0):** Gross margin precentage has almost zero correlation with Quantity (0.011).


1. **Branch C** has highest sales compared to Branch A and B
2. **Food & beverages** is top selling product line  
3. **Ewallet & Cash** are most used payment methods ~34% each



After analyzing retail sales data, 3 key findings:
1. Quantity is the main driver of Sales - stock management is critical
2. Tax 5%, Sales, cogs are calculated fields with perfect correlation  
3. Discount has negative impact on profit - need pricing strategy
